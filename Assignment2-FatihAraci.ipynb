### Logistic Regression

Using the Iris dataset, find the optimum logistic regression parameters for the following three problems:

Classify 1 versus 2,3
Classify 2 versus 1,3
Classify 3 versus 1,2
Your program should output three parameter vectors, each of lenght 4

```python
# Lineer Regresyon
# Sürekliliği olan değerleri bulmak icin kullanılır (yas - gelir)
# daha çok sayılar için geçerlidir


# Logistic Regresyon
# Daha çok kategori değişikliklerini bulmak için
# Sınıflandırma vs. gibi
# binary outcome
# sürekli -1 ve 1 arasında cıkıyor ona göre sınıflandırma yapıyorsun
```

```python
%matplotlib inline

import scipy as sc
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt


# Iris dataset
iris = pd.read_csv("data/iris.txt", sep=" ")

iris
```

```python
# sececegimiz cicekleri 1 geri kalanları 0 yaparız ki ayrım yapabilelim
# bu nedenle c colonu (yani cicek tipi) 1 den buyuk olanları sıfıra esitleyerk
# tip sayımızı 2'ye dusuyoruz


# logistic function tanımlayalım
# bunun sebebi logistic regression 2 adet yaklasımımız var:
#    - olasılık daima 0 ve 1 arasında olmalı
#    - hiç bir zaman negatif çıkmamalı
# bu sartları saglamak icin bir fonksiyon yazıyoruz bu nednele logistic regressionicin bu fonk yazılır

# Define X matrix with the feature from the iris dataset
sl = np.vstack(iris.sl)
sw = np.vstack(iris.sw)
pl = np.vstack(iris.pl)
pw = np.vstack(iris.pw)
c  = np.vstack(iris.c)

# features
features = np.hstack((sl,sw,pl,pw))

def sigmoid(t):
    return 1 / ( 1 + np.exp(-t))


# oncelikle elimizdeki dataseti kendi seçtiğimiz bir w olan birim matris ile carpıyoruz
# daha sonra buldugumuz deger ile sigmoid fonksiyonunu kullanarak 0 ve 1 arasına yerleştiriyoruz
# bu sekilde ikinci ve sonraki w degerlerini bulabilir hale geliyorz
# eta ve itration degerlerini kullnarak en optimum w'yu bulmaya calısıyoruz
# optimum w'yu bulmak için gradient ascend kullanılır

# Define eta ve iteration
eta = 0.001 
iteration=10000

# x.T[y-sigmoid(x*w)]
# Define classification function to find the w matrix from the equality y = Ax + w
def classify(category, eta = 0.001):
    
    # Define y matrix
    y = np.vstack(iris.c)
    
    # Define w* as a dummy for now
    w = np.matrix(np.ones(4)).T
    
    # Group by category (1 for the class that will be classified with, 0 for the others)
    for i in range(len(y)):
        y[i,0] = 1 if y[i,0] == category else 0
        
    # Iterate to find the optimal w matrix
    for i in range(10000):
        prob = sigmoid(features * w)
        # Gradient ascent as a general form: x.T * (y - sigmoid(x * w)) = d L / d w
        w = w + eta * features.T * (y - prob)
    
    print ("min value of category:", category ,"]", np.min(features * w))   
    print ("max value of category:", category ,"]", np.max(features * w))
    print ("----------------------")
    return w
```

```python
# cicekleri classify etmek icin gerekli w degerini hesaplıyoruz
# daha sonra buradaki degeleri sigmoid e göndererek 0 ile 1 arasına alacagız
# 0 ile 1 arasında bir sart secerek ornegin 0.5'den buyuk ya da kucuk olacak sekilde
# sınıflandırma yapacagız

setosaModel = classify(1)
versicolorModel = classify(2)
virginicaModel = classify(3)

def classify_result(features, flower_type):
    output = sigmoid(features * flower_type)
    output[ output <  0.5 ] = 0
    output[ output >= 0.5 ] = 1
    return output

# buldugumuz optimum w degerlerini kullanarak cicekleri classify edelim.
# boylece c kolonu ile kıysalama yaparak basarı oranımızı hesaplayabiliriz.
setosa_result     = classify_result(features, setosaModel)
versicolor_result = classify_result(features, versicolorModel)
virginica_result  = classify_result(features, virginicaModel)
```
