```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pylab as plt
import pandas as pd

def sigmoid(x):
    return 1/(1+np.exp(-x))


def w_calculator(Features,y,eta):
    w = np.mat('[0;0;0;0]')
    for i in range(10000):
        pr = sigmoid(Features * w)
        w = w + eta * Features.T * (y - pr)    
    return w

def test(w_n):
    print("-------results for values of 1--------")
    for i in range (50):
        a = sl[i]
        b = sw[i]
        c = pl[i]
        d = pw[i]
        Feature = np.hstack([a,b,c,d])
        print(Feature*w_n)
    print("-------results for values of 2--------")
    for i in range (51,100):
        a = sl[i]
        b = sw[i]
        c = pl[i]
        d = pw[i]
        Feature = np.hstack([a,b,c,d])
        print(Feature*w_n)
    print("-------results for values of 3--------")
    for i in range (101,150):
        a = sl[i]
        b = sw[i]
        c = pl[i]
        d = pw[i]
        Feature = np.hstack([a,b,c,d])
        print(Feature*w_n)
```

```python
df_iris = pd.read_csv(u'data/iris.txt',sep=' ')
```

```python
sl = np.matrix(df_iris.sl[:]).T
sw = np.matrix(df_iris.sw[:]).T
pl = np.matrix(df_iris.pl[:]).T
pw = np.matrix(df_iris.pw[:]).T
FeatureVector = np.hstack((sl,sw,pl,pw))
```

### Classify 1 versus 2,3

```python
y = np.matrix(df_iris.c[:]).T
y[y > 1] = 0
w1 = w_calculator(FeatureVector,y,0.001)
print(w1)
```


### Test-1:

```python
test(w1)
```

### Result: Iris of type 1 can distinguish because it has positive outcomes

### Classify 2 versus 1,3

```python
y = np.matrix(df_iris.c[:]).T
y[y == 1] = 0
y[y == 2] = 1
y[y == 3] = 0
w2 = w_calculator(FeatureVector,y,0.0001)
print(w2)
```

## Test-2 :

```python
test(w2)
```

### Result: Iris of type 2 is not easy to distinguish because of mixed outcomes.

### Classify 3 versus 1,2

```python
y = np.matrix(df_iris.c[:]).T
y[y < 3] = 0
y[y == 3] = 1
w3 = w_calculator(FeatureVector,y,0.001)
print(w3)
```

## Test-3:

```python
test(w3)
```

### Result: Iris of type 3 can distinguish because it has mostly positive outcomes.
